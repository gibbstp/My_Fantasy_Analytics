```python
# setting root directory for glob selections
root_dir = "../../db/data/"

matching_string = "*.csv"
files = glob.glob(matching_string, root_dir=root_dir)
```

```python
def return_file_week(file_name):
    """ 
    funtion to process file name string to return name of week
    
    Parameters:
    -----------
    file name: str
        name of file to be processed
    
    Return:
    -----------
    str
        week number from file name    
    """
    
    list_file_name = file_name.split("_")

    str_wk_num = list_file_name[2].split(".")[0]

    return str_wk_num
```

```python
def iterative_read_csv_to_pandas(files):
    
    table_names = [return_file_week(file) for file in files]
    
    list_of_dataframes = []
    
    for table_name, file in zip(table_names, files):
        
        df = pd.read_csv(f'{root_dir}{file}', index_col=0, header=0)
        
        list_of_dataframes.append(df)
        
    return list_of_dataframes
```
